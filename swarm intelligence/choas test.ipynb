Swarm intelligence  

```python
import matplotlib.pyplot as plt
import numpy as np
p1 = []
p2 = []
def test_chaos(x, y, n=4, i =100):
    if i <=0 :
        return 
    f1 = n*x*(1-x)
    f2 = x+y
    if f2>1.0:
        f2 = f2-1
    p1.append(f1)
    p2.append(f2)
    return test_chaos(f1, f2, i=i-1)
test_chaos(0.8,0.775)  
p1 = np.array(p1)
p2 = np.array(p2)
plt.plot(p1,p2)
plt.show()  
```
